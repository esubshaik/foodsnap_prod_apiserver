```python
import csv
food_data = []
with open('food_data.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        food_data.append(row)

# Calculate the user's Daily Caloric Needs (DCN)
weight = 90
height = 165
age = 21
gender = 'M'

if gender.lower() == "m":
    DCN = 66.5 + (13.75 * weight) + (5.003 * height) - (6.755 * age)
else:
    DCN = 655 + (9.563 * weight) + (1.850 * height) - (4.676 * age)

# Determine the User's BMI Category
BMI = weight / (height / 100)**2
BMI_category = ""
if BMI < 18.5:
    BMI_category = "Underweight"
elif BMI >= 18.5 and BMI < 25:
    BMI_category = "Normal Weight"
elif BMI >= 25 and BMI < 30:
    BMI_category = "Overweight"
else:
    BMI_category = "Obese"

# Suggest Foods Based on BMI Category
suggested_foods = []
for food in food_data:
    if BMI_category == "Underweight":
        if float(food['PROTEIN(G)']) > 10 and float(food['CARBOHYDRATES(G)']) > 10 or float(food['CALORIES(G)']) > 200:
            suggested_foods.append(food['FOOD'])
    elif BMI_category == "Normal Weight":
        if float(food['PROTEIN(G)']) > 10 and float(food['CARBOHYDRATES(G)']) > 10 and float(food['FAT(G)']) < 10 or float(food['CALORIES(G)']) >= 10:
            suggested_foods.append(food['FOOD'])
    elif BMI_category == "Overweight" or BMI_category == "Obese":
        if float(food['PROTEIN(G)']) > 10 and float(food['CARBOHYDRATES(G)']) < 10 and float(food['FAT(G)']) < 10 or float(food['CALORIES(G)']) < 200:
            suggested_foods.append(food['FOOD'])

# Print the suggested foods
print("Suggested Foods for {}:".format(BMI_category))
for food in suggested_foods:
    print("* {}".format(food))
```

```python
import csv

# Read the CSV file containing food data
with open('food_data.csv', 'r') as f:
    reader = csv.reader(f)
    food_data = list(reader)

# Get the user's BMI
bmi = float(input("Enter your BMI: "))

# Create a list to store suggested food items
suggested_foods = []

# Iterate through the food data and filter based on BMI
for food in food_data:
    if bmi < 18.5:
        # Suggest high-calorie, high-fat, and high-protein foods
        if float(food[1]) >= 250 and float(food[3]) >= 10 and float(food[4]) >= 10:
            suggested_foods.append(food[0])
    elif bmi >= 18.5 and bmi < 25:
        # Suggest balanced diet foods
        suggested_foods.append(food[0])
    elif bmi >= 25 and bmi < 30:
        # Suggest low-calorie, high-fiber, and high-nutrient foods
        if float(food[1]) <= 200 and float(food[5]) >= 5 and float(food[2]) >= 5:
            suggested_foods.append(food[0])
    elif bmi >= 30:
        # Suggest very low-calorie, high-fiber foods
        if float(food[1]) <= 150 and float(food[5]) >= 10:
            suggested_foods.append(food[0])

# Print the suggested food items
print("Suggested food items for your BMI:")
for food in suggested_foods:
    print("- {}".format(food))
```
